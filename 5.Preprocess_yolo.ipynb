```python
import pandas as pd
import shutil
import os
import numpy as np
from sklearn.model_selection import train_test_split
import cv2
```

```python
df = pd.read_csv('train_detect_mask.csv',header=None)
```

```python
df.columns = ['path','x1','y1','x2','y2','class']
```

```python
files = np.unique(df['path'])
```

```python
train,val = train_test_split(files,test_size=0.2,random_state=1)
```

```python
for idx in files:
    print(idx)
    name = os.path.basename(idx)
    shutil.copy2(idx,'data_y/test/{}'.format(name))
```

```python
traindf = df[df['path'].isin(train) ]
valdf = df[df['path'].isin(val) ]
```

```python
name = os.path.basename(traindf.iloc[0]['path'])
num = 0
for i in range(len(traindf)):
    data = traindf.iloc[i]
    path = data['path'] 
    new_name = os.path.basename(path)
    if new_name == name:
        num += 1
    else:
        num = 0
    print(new_name, num)
    img = cv2.imread(path)
    h, w, c = img.shape
    P_x = float(data[1]+data[3]) / (2*w)
    P_y = float(data[2]+data[4]) / (2*h)
    P_w = (float(data[3]-data[1]) / w)
    P_h = (float(data[4]-data[2]) / h)
    with open('data_y/labels/train/{}'.format(new_name.replace(".png",".txt")), 'a') as f:
        if np.isnan(P_x) != True:
            f.write("{} {} {} {} {}\n" .format(0, str(P_x), str(P_y), str(P_w), str(P_h)))
        f.close()
    name = new_name
```
