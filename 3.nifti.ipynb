```python
import nibabel as nib
import cv2
import glob
import os
```

```python
for path in glob.glob('ROI/Add/*.nii.gz'):
    proxy = nib.load(path)
    header = proxy.header
    arr = proxy.get_fdata()
    arr = cv2.transpose(arr)
    print(arr.shape)    
    savepath = path.replace(".nii.gz","_mask.png")
    cv2.imwrite(savepath,arr)


```
