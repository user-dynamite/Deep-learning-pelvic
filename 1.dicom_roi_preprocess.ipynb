```python
import pandas as pd
import numpy as np
import pydicom
from read_roi import read_roi_file
from read_roi import read_roi_zip
import os
import glob
from matplotlib import pyplot as plt
from pydicom.pixel_data_handlers.util import apply_modality_lut, apply_voi_lut
from PIL import Image
import cv2
```

```python
dicoms = glob.glob("dataset/*/*.dcm")
```

```python
dicoms
```

```python
for dicom in dicoms:
    print(dicom)
    ds = pydicom.read_file(dicom)
    arr = ds.pixel_array
    image_2d = arr.astype(float)
    image_2d_scaled = (np.maximum(image_2d,0) / image_2d.max()) * 255.0
    image_2d_scaled = np.uint8(image_2d_scaled)
    if ds.PhotometricInterpretation != 'MONOCHROME2':
        image_2d_scaled = np.invert(image_2d_scaled)
    image = Image.fromarray(image_2d_scaled)
    image.save(dicom.replace('dcm','jpg'))
```

```python
for dicom in dicoms:
    print(dicom)
    ds = pydicom.read_file(dicom)
    if(('RescaleSlope' in ds) and ('RescaleIntercept' in ds)):
        print('oh?')
        rs = ds.RescaleSlope
        ri = ds.RescaleIntercept
        arr = (ds.pixel_array * rs) + ri
    else:
        arr = ds.pixel_array
    try:
        arr = apply_modality_lut(arr, ds)
        arr = apply_voi_lut(arr, ds)
    except:
        print("nope")
    img_norm2 = cv2.normalize(arr, None, 0, 255, cv2.NORM_MINMAX)
    img_norm2 = img_norm2.astype(np.uint8)
    if ds.PhotometricInterpretation != 'MONOCHROME2':
        img_norm2 = np.invert(img_norm2)
    cv2.imwrite(dicom.replace('dcm','png'),img_norm2)
```

```python
for path in glob.glob("dataset/*/*[!mask,!HE,!box].png"):
    print(path)
    img = cv2.imread(path,0)
    clahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(8,8))
    img2 = clahe.apply(img)
    sv_path = path.replace('.png','_HE.png')
    cv2.imwrite(sv_path, img2)
```

```python
for path in glob.glob("dataset/*/*[!mask,!HE, !box].png"):
    maskpath = path.replace(".png","_mask.png")
    img = cv2.imread(path)
    mask = cv2.imread(maskpath)
    try:
        crop_img = np.where(mask != 0,img,0)
#         savepath = path.replace("_HE","")
        cv2.imwrite(path,crop_img)
    except:
        print(path)
        continue
```

```python
df= []
for pngf in pngs:
    name = pngf.split("/")[2]
    try:
        roi = read_roi_file(pngf)
        R_val = list(roi.values())[0]
        x1 = R_val['left']
        x2 = x1+R_val['width']
        y1 = R_val['top']
        y2 = y1+R_val['height']
        fname = pngf
        df.append([name,x1,y1,x2,y2,'fracture',name])
        print(pngf,R_val['name'])
    except:
        try:
            roi = read_roi_zip(pngf.replace('png','zip'))
            all_val = list(roi.values())
            for R_val in all_val:
                x1 = R_val['left']
                x2 = x1+R_val['width']
                y1 = R_val['top']
                y2 = y1+R_val['height']
                fname = pngf
                df.append([name,x1,y1,x2,y2,'fracture',name])
                print(pngf,R_val['name'])
        except:
            df.append([name,None,None,None,None,None,None])
            print('C')
```
