```python
import glob
import pandas as pd
import pydicom
```

```python
df = pd.read_csv("Retinanet_evaluation_log_합침3.csv")
```

```python
def padding(img):

    try:
        h,w = img.shape
    except:
        print('파일을 확인후 다시 시작하세요.')
        raise

    if h < w:
        set_size = w
        new_width = set_size
        new_height = int(new_width * (h/w))
    else:
        set_size = h
        new_height = set_size
        new_width = int(new_height * (w/h))

    try:
        h,w = img.shape
    except:
        print('파일을 확인후 다시 시작하세요.')
        raise

    delta_w = set_size - w
    delta_h = set_size - h
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)

    return top,left
```

```python
w = []
h = []
x1 = []
x2 = []
y1 = []
y2 = []
for idx,row in df.iterrows():
    name = row[0]
    print(name)
    for path in glob.glob("{}".format(name)):
        dcm = path.replace(".png",".dcm")
        ds = pydicom.read_file(dcm)      
        w.append(ds.Columns)
        h.append(ds.Rows)
        top,left = padding(ds.pixel_array)
    x1.append(row[6])
    y1.append(row[7])
    x2.append(row[8])
    y2.append(row[9])
```

```python
df['width'] = w
df['height'] = h
df['x1'] = x1
df['x2'] = x2
df['y1'] = y1
df['y2'] = y2
```
