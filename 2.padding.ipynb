```python
import glob
import cv2
import numpy as np
import shutil
import pydicom
from matplotlib import pyplot as plt
import os
```

```python
def padding(img, set_size):

    try:
        h,w,c = img.shape
    except:
        print('파일을 확인후 다시 시작하세요.')
        raise

    if h < w:
        new_width = set_size
        new_height = int(new_width * (h/w))
    else:
        new_height = set_size
        new_width = int(new_height * (w/h))

    if max(h, w) < set_size:
        img = cv2.resize(img, (new_width, new_height), cv2.INTER_CUBIC)
    else:
        img = cv2.resize(img, (new_width, new_height), cv2.INTER_AREA)
 
    try:
        h,w,c = img.shape
    except:
        print('파일을 확인후 다시 시작하세요.')
        raise

    delta_w = set_size - w
    delta_h = set_size - h
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)

    new_img = cv2.copyMakeBorder(img, top, bottom, left, right, cv2.BORDER_CONSTANT, value=[0, 0, 0])

    return new_img

def unpadding(img,width,height):
    
    set_size = np.max([width,height])
    print(set_size)
    h,w,c = img.shape
    
    if height < width:
        new_width = set_size
        new_height = int(new_width * (height/width))
    else:
        new_height = set_size
        new_width = int(new_height * (width/height))
    print(new_width,new_height)
    if max(h, w) < set_size:
        img = cv2.resize(img, (set_size, set_size), cv2.INTER_CUBIC)
    else:
        img = cv2.resize(img, (set_size, set_size), cv2.INTER_AREA)
        
    h,w,c = img.shape
    if width==height:
        return img
    
    delta_w = set_size - width
    delta_h = set_size - height
    print(delta_w,delta_h)
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)

    new_img = img[top:h-(bottom),left:w-(right)]
    
    return new_img
```

```python
num = 0
for img in glob.glob("6_result/*/pred_total/*/*.png"):
    name = os.path.basename(img)
#     dcm = img.replace(dirname,'dataset')
    dcmname = name.replace("_HE.png",".dcm")
    dcm = glob.glob('dataset/*/{}'.format(dcmname))[0]
    img_arr = cv2.imread(img)
    df = pydicom.read_file(dcm)
    W = df.Columns
    H = df.Rows
    mask = unpadding(img_arr,W,H)
    ori_img = cv2.imread(dcm.replace(".dcm",'.png'))
    print(mask.shape)
    print(ori_img.shape)
#     added_image = cv2.addWeighted(mask,0.5,ori_img,0.5,0)
    cv2.imwrite(dcm.replace(".dcm",'_mask.png'),mask)
#     plt.imshow(added_image)
#     plt.show()
    num += 1
print(num)
```
