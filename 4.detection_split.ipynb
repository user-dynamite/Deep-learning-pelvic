```python
import pandas as pd
import numpy as np
import shutil
import os
import glob
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv('Data_select.csv')
```

```python
df_name = df.loc[:,['path','grade']]
```

```python
df_name = df_name.drop_duplicates()
```

```python
S_kfold = StratifiedKFold(n_splits=4, random_state=5, shuffle=True)
```

```python
for i, (train_index, test_index) in enumerate(S_kfold.split(df_name, df_name['grade'])):
    CV_train = df_name.iloc[train_index]
    CV_test = df_name.iloc[test_index]
    df_train = pd.merge(df, CV_train, left_on='path', right_on='path', how='inner')
    df_test = pd.merge(df, CV_test, left_on='path', right_on='path', how='inner')
    df_train.to_csv('train_{}.csv'.format(i),index=False,header=False)
    df_test.to_csv('test_{}.csv'.format(i),index=False,header=False)
    print(len(df_train),len(df_test))
```
